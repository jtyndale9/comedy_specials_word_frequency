```python
import nltk
from pathlib import PurePath
from nltk.corpus import stopwords
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = PurePath(r'C:\Users\john4\Documents\Software Development\Comedy Data Viz\Comedy_Transcripts')
#print("Path: " , path)
file = r'Ronny_Chieng_destroys_america.txt'
file1 = path / file
print("Path: \n", file1)
```

```python
file1_read = open(file1, encoding='utf-8')
file1_data = file1_read.read()
#print(file1_data) #this has the whole text file raw data saved in it

tokens = nltk.word_tokenize(file1_data)
#print("tokens: \n", tokens) # this contains all messy tokens, capitals, and non characters like '['
tokens=[word.lower() for word in tokens if word.isalpha()] #change everything to lowercase
#print("tokens: \n", tokens) # this contains all cleaned up lower case words
fdist1 = nltk.FreqDist(tokens)
print(fdist1)
```

```python
filtered_word_freq = dict((word, freq) for word, freq in fdist1.items() if not word.isdigit())
#print(filtered_word_freq) # This contains the amount of times a word appears in the token list
print("type: ", type(filtered_word_freq))
```

```python
for key, value in filtered_word_freq.items():
    print(key, '->', value)
```

```python
# Sort it so you get the most commonly used words on top
sorted_dict = sorted(filtered_word_freq, key=filtered_word_freq.get, reverse=True)
type(sorted_dict)
sorted_dict
```

```python
sorted_values = sorted(filtered_word_freq.values(), reverse=True) # Sort the values
sorted_dict = {}

for i in sorted_values:
    for k in filtered_word_freq.keys():
        if filtered_word_freq[k] == i:
            sorted_dict[k] = filtered_word_freq[k]
```

```python
for key, value in sorted_dict.items():
    print(key, '->', value)
```

```python
#print(stopwords.words('english'))
stop_words = stopwords.words('english')
stop_words
```

```python
# Take all stop words out of the data
new_dict = {}

for key, value in sorted_dict.items():
    if key not in stop_words:
        new_dict[key] = value
```
