```python
import nltk
from pathlib import PurePath
```

```python
path = PurePath(r'C:\Users\john4\Documents\Comedy_Transcripts')
path
```

```python
file = r'Bill_Burr_2022.txt'
```

```python
file1 = path / file
file1
```

```python
file1_read = open(file1, encoding='utf-8')
```

```python
file1_data = file1_read.read()
#file1_data
```

```python
tokens = nltk.word_tokenize(file1_data)
#tokens
```

```python
tokens=[word.lower() for word in tokens if word.isalpha()]
#tokens
```

```python
fdist1 = nltk.FreqDist(tokens)
fdist1
```

```python
filtered_word_freq = dict((word, freq) for word, freq in fdist1.items() if not word.isdigit())
#filtered_word_freq
```

```python
type(filtered_word_freq)
```

```python
for key, value in filtered_word_freq.items():
    print(key, '->', value)
```

```python
sorted_dict = sorted(filtered_word_freq, key=filtered_word_freq.get, reverse=True)
type(sorted_dict)
```

```python
sorted_values = sorted(filtered_word_freq.values(), reverse=True) # Sort the values
sorted_dict = {}

for i in sorted_values:
    for k in filtered_word_freq.keys():
        if filtered_word_freq[k] == i:
            sorted_dict[k] = filtered_word_freq[k]
```

```python
for key, value in sorted_dict.items():
    print(key, '->', value)
```
